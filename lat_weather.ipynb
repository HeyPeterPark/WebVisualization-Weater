## Weater Data Analysis for Website

```python
import matplotlib.pyplot as plt
from matplotlib import style
import pandas as pd
import numpy as np
import time

#style.use('fivethirtyeight')
# style.use('ggplot')
style.use('seaborn')

temp_df = pd.read_csv("Resources/cities.csv")
temp_df.set_index("City_ID", inplace = True,)

temp_df.head(20)
```

#### Latitude vs. Temperature Plot

```python
x_axis = temp_df["Lat"]

plt.scatter(x_axis, temp_df["Max Temp"], marker = "o", s = 35, edgecolor = "k")

plt.title(f"City Latitude vs. Max Temperature (01/05/17)")
plt.ylabel("Max. Temperature (F)")
plt.xlabel("Latitude")
plt.axvline(x = 0, c = "r", alpha = 0.75, ls = "dotted")
plt.grid(alpha = 0.75)

plt.savefig("Resources/assets/lat_v_temp.png", bbox_inches = "tight")

plt.show()
```

#### Latitude vs. Humidity Plot

```python
plt.scatter(x_axis, temp_df["Humidity"], marker = "o", s = 35, edgecolor = "k")

plt.title(f"City Latitude vs. Humidity (01/05/17)")
plt.ylabel("Humidity (%)")
plt.xlabel("Latitude")
plt.axvline(x = 0, c = "r", alpha = 0.75, ls = "dotted")
plt.grid(alpha = 0.75)

plt.savefig("Resources/assets/lat_v_humidity.png", bbox_inches = "tight")

plt.show()
```
